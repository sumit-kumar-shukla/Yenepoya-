```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_html('/content/NIA Cases _ National Investigation Agency.html')
```

```python
df[1]
```

```python
NIA_DLI = df[0]
NIA_GUW = df[1]
NIA_HYD = df[2]
NIA_KOC = df[3]
NIA_MUM = df[4]
NIA_LKW = df[5]
NIA_JMU = df[6]
NIA_RPR = df[7]
NIA_RNC = df[8]
NIA_IMP = df[9]
NIA_CHN = df[10]
NIA_AHM = df[11]
NIA_JPR = df[12]
```

```python
nic_cases = pd.concat([NIA_DLI, NIA_GUW, NIA_HYD, NIA_JMU, NIA_KOC, NIA_LKW, NIA_MUM, NIA_RNC,
                       NIA_RPR, NIA_IMP, NIA_CHN, NIA_AHM, NIA_JPR], axis = 0, ignore_index=True)
```

```python
nic_cases.sample()
```

```python
# ADDING COLS
nic_cases['Branch'] = nic_cases['Case No'].apply(lambda x: x[-3:])
nic_cases.sample(5)
```

```python
values = ['DLI', 'GUW', 'HYD', 'JMU', 'KOC', 'LKW', 'MUM', 'RNC', 'RPR', 'IMP', 'CHE', 'AMD', 'JPR']
to_change = ['Delhi','Guwahati' ,'Hyderabad' ,'Jammu' , 'Kochi', 'Lucknow' , 'Mumbai' , 'Ranchi', 'Raipur','Imphal','Chennai', 'Ahemdabad', 'Jaipur']

#changing names of values to to_change
nic_cases['Branch'].replace(values,to_change,inplace=True)
```

```python
# removing null data from Orgaization
# replacing "-" with unknown
nic_cases['Organisation / Gang'].replace('-','Undiscovered', inplace=True)
```

```python
nic_cases.to_csv('NIC_uptd.csv', index=False)
```
